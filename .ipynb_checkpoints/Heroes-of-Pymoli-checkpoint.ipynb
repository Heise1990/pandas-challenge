```python
import pandas
from pandas import DataFrame
```

```python
from IPython.display import display
pandas.options.display.float_format = '{:,.2f}'.format
```

```python
path = "Heroes-of-Pymoli/Resources/purchase_data.csv"
```

```python
pymoliPurchases = pandas.read_csv(path)
```

```python
pymoliPurchases
```

```python
count = pymoliPurchases["SN"].value_counts()

playerCount_df = pandas.DataFrame({
    
    "Total Players": [len(count)]
})

playerCount_df
```

```python
countID = pymoliPurchases["Item ID"].value_counts()
averagePrice = pymoliPurchases["Price"].mean()
numberPurchases = len(pymoliPurchases)
totalRevenue = pymoliPurchases["Price"].sum()
```

```python
purchasingAnalysis_df = pandas.DataFrame({
    
    "Number of Unique Items": [len(countID)],
    "Average Price": [averagePrice],
    "Number of Purchases": [numberPurchases],
    "Total Revenue": [totalRevenue]
    
})

purchasingAnalysis_df["Average Price"] = purchasingAnalysis_df["Average Price"].map("${:.2f}".format)
purchasingAnalysis_df["Total Revenue"] = purchasingAnalysis_df["Total Revenue"].map("${:,.2f}".format)
purchasingAnalysis_df
```

```python
pyNoDupe = pymoliPurchases.drop_duplicates("SN")
```

```python
genderBreakdown = pyNoDupe["Gender"].value_counts()
genderCount = pyNoDupe.count()[0]
genderPrcnt = genderBreakdown/genderCount*100
genderBreakdown

genderBreakdown_df = pandas.DataFrame({
    
    "Total Count": genderBreakdown,
    "Percentage of Players": genderPrcnt})

genderBreakdown_df
```

```python
purchaseCount = pymoliPurchases.groupby(["Gender"]).count()["Price"]
avgPurchasePrice = pymoliPurchases.groupby(["Gender"]).mean()["Price"]
totalPurchaseValue = pymoliPurchases.groupby(["Gender"]).sum()["Price"]
AvgPurchasePerPerson = pymoliPurchases.mean()["Price"]

genPurchaseAnalysis_df = pandas.DataFrame({
    
    "Purchase Count": purchaseCount,
    "Average Purchase Price": avgPurchasePrice,
    "Total Purchase Value": totalPurchaseValue,
    "Avg Total Purchase per Person": totalPurchaseValue/genderBreakdown
    
})

genPurchaseAnalysis_df["Average Purchase Price"] = genPurchaseAnalysis_df["Average Purchase Price"].map("${:.2f}".format)
genPurchaseAnalysis_df["Total Purchase Value"] = genPurchaseAnalysis_df["Total Purchase Value"].map("${:,.2f}".format)
genPurchaseAnalysis_df["Avg Total Purchase per Person"] = genPurchaseAnalysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

genPurchaseAnalysis_df
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
ageBins = pyNoDupe.loc[:,["SN","Age"]]
ageBins["Age Bins"] = pandas.cut(ageBins["Age"], bins, labels=group_names)
ageBins

totalAgeCount = ageBins["Age Bins"].value_counts()
prcntAgeRange = (totalAgeCount/genderCount * 100)

ageDemo_df = pandas.DataFrame({
    "Total Count": totalAgeCount,
    "Percentage of Players": prcntAgeRange

})

ageDemo_df["Percentage of Players"] = ageDemo_df["Percentage of Players"].map("{:.2f}%".format)

ageDemo_df.sort_index()
```

```python
purchaseBins = pymoliPurchases.loc[:,["SN","Age", "Price"]]
purchaseBins["Age Bins"] = pandas.cut(purchaseBins["Age"], bins, labels=group_names)
purchaseBins

totalPurchaseCount = purchaseBins["Age Bins"].value_counts()
avgPurchasePrice = purchaseBins.groupby(["Age Bins"]).mean()["Price"]
totalPurchaseValue = purchaseBins.groupby(["Age Bins"]).sum()["Price"]
avgTotalPurchases = (totalPurchaseValue/ageBins["Age Bins"].value_counts())

purchaseBins_df = pandas.DataFrame({
    "Purchase Count": totalPurchaseCount,
    "Average Purchase Price": avgPurchasePrice,
    "Total Purchase Value": totalPurchaseValue,
    "Avg Total Purchase per Person": avgTotalPurchases

    
})

purchaseBins_df["Average Purchase Price"] = purchaseBins_df["Average Purchase Price"].map("${:.2f}".format)
purchaseBins_df["Total Purchase Value"] = purchaseBins_df["Total Purchase Value"].map("${:,.2f}".format)
purchaseBins_df["Avg Total Purchase per Person"] = purchaseBins_df["Avg Total Purchase per Person"].map("${:.2f}".format)

purchaseBins_df.sort_index()
```

```python
purchaseCount = pymoliPurchases.groupby(["SN"]).count()["Price"]
avgPurchPrice = pymoliPurchases.groupby(["SN"]).mean()["Price"]
tPV = pymoliPurchases.groupby(["SN"]).sum()["Price"]

purchasingAnalysis_df = pandas.DataFrame({
    "Purchase Count": purchaseCount,
    "Average Purchase Price": avgPurchPrice,
    "Total Purchase Value": tPV
    
})

purchasingAnalysis_df["Average Purchase Price"] = purchasingAnalysis_df["Average Purchase Price"].map("${:.2f}".format)

purchasingAnalysis_df.sort_values("Total Purchase Value",ascending = False).head() 
```
